```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_runs = 2000
n_steps = 10000
q_star_all = np.random.normal(0, 1, 10)
```

```python
epsilon = 0
step_optimal_action_count_1 = np.zeros(n_steps)
for i in range(n_runs):
    q_star = np.copy(q_star_all)
    q = np.ones(10) * 5
    n = np.zeros(10)
    for j in range(n_steps):
        if np.random.random() < epsilon:
            choice = np.random.randint(10)
        else:
            choice = np.argmax(q)
        if choice == np.argmax(q_star):
            step_optimal_action_count_1[j] += 1
        reward = np.random.normal(q_star[choice], 1)
        n[choice] += 1
        q[choice] += (reward - q[choice]) / n[choice]
```

```python
epsilon = 0.1
step_optimal_action_count_2 = np.zeros(n_steps)
for i in range(n_runs):
    q_star = np.copy(q_star_all)
    q = np.zeros(10)
    n = np.zeros(10)
    for j in range(n_steps):
        if np.random.random() < epsilon:
            choice = np.random.randint(10)
        else:
            choice = np.argmax(q)
        if choice == np.argmax(q_star):
            step_optimal_action_count_2[j] += 1
        reward = np.random.normal(q_star[choice], 1)
        n[choice] += 1
        q[choice] += (reward - q[choice]) / n[choice]
```

```python
step_optimal_action_count_1 /= (n_runs / 100)
step_optimal_action_count_2 /= (n_runs / 100)
```

```python
plt.plot(step_optimal_action_count_1)
plt.plot(step_optimal_action_count_2)
plt.legend(['Optimistic Greedy', 'Realistic, e-greedy'], loc='upper left')
plt.xlabel('Steps')
plt.ylabel('Optimal Action (%)')
```

# Observations
Initially, all of the rewards are lower than the agents' Q-values which results in each of the agents getting disappointed. As the decrease in the reward for the optimal action is lower compared to the others, the agents have a higher chance of selecting the optimal action which leads to the kink. However, again the agents are disappointed leading to them exploring other options again. This leads to the initial oscillations upto the point the Q-values reach somewhere around the Q* values.
