```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n_runs = 2000
n_steps = 10000
epsilon = 0.1
step_size = 0.1
```

```python
step_reward_sum_1 = np.zeros(n_steps)
step_optimal_action_count_1 = np.zeros(n_steps)
for i in range(n_runs):
    q_star = np.zeros(10)
    q = np.zeros(10)
    n = np.zeros(10)
    for j in range(n_steps):
        q_star += np.random.normal(0, 0.01, 10)
        if np.random.random() < epsilon:
            choice = np.random.randint(10)
        else:
            choice = np.argmax(q)
        if choice == np.argmax(q_star):
            step_optimal_action_count_1[j] += 1
        reward = np.random.normal(q_star[choice], 1)
        step_reward_sum_1[j] += reward
        n[choice] += 1
        q[choice] += (reward - q[choice]) / n[choice]
```

```python
step_reward_sum_2 = np.zeros(n_steps)
step_optimal_action_count_2 = np.zeros(n_steps)
for i in range(n_runs):
    q_star = np.zeros(10)
    q = np.zeros(10)
    n = np.zeros(10)
    for j in range(n_steps):
        q_star += np.random.normal(0, 0.01, 10)
        if np.random.random() < epsilon:
            choice = np.random.randint(10)
        else:
            choice = np.argmax(q)
        if choice == np.argmax(q_star):
            step_optimal_action_count_2[j] += 1
        reward = np.random.normal(q_star[choice], 1)
        step_reward_sum_2[j] += reward
        n[choice] += 1
        q[choice] += (reward - q[choice]) * step_size
```

```python
step_reward_sum_1 /= n_runs
step_optimal_action_count_1 /= (n_runs / 100)
step_reward_sum_2 /= n_runs
step_optimal_action_count_2 /= (n_runs / 100)
```

```python
plt.plot(step_reward_sum_1)
plt.plot(step_reward_sum_2)
plt.legend(['Sample averages', ('Step size = ' + str(step_size))], loc='upper left')
plt.xlabel('Steps')
plt.ylabel('Average Reward')
```

```python
plt.plot(step_optimal_action_count_1)
plt.plot(step_optimal_action_count_2)
plt.legend(['Sample averages', ('Step size = ' + str(step_size))], loc='upper left')
plt.xlabel('Steps')
plt.ylabel('Optimal Action (%)')
```
